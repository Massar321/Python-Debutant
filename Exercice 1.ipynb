```python
nom_client = "Massar"
prix_article = 15.99 
quantite = 3 
est_fidel = True  

#calculer le total brut de l'arcticle du client massar 
total = prix_article * quantite 
print(f"le prix total du client {nom_client} : {total} $")

#calculer le total après remise 
remise = 0.10
total_remise = total *(1-remise)
if est_fidel : 
    print(f"le prix total après remise pour {nom_client} : {total_remise : .2F} $")
else : 
    print(f"pas de prix total pour  {nom_client} : {total_remise : .2F} $")
```
